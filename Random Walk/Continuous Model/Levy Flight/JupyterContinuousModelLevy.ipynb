```python
from RandomWalkContinuousLevy import *
import matplotlib.pyplot as plt
import numpy as np
from mesa.batchrunner import BatchRunner

def displacement(model):
    return model.grid.get_distance((50,50), (model.schedule.agents[0].pos))

# Larger world size due to increased step length, random walk in any direction
sim = WorldModel(1, 1000, 1000, 2*np.pi)
```

```python
sim.schedule.agents[0].pos
for i in range(100):
    sim.step()
```

```python
data = sim.data_collector.get_agent_vars_dataframe()
x,y = [],[]
for i in data["Position"]:
    x.append(i[0])
    y.append(i[1])
finalx, finaly = x[-1], y[-1]
plt.plot(x,y)
plt.scatter(finalx, finaly, c="r", marker="x")
plt.grid(True, linewidth=1)
plt.xlim(0,1000)
plt.ylim(0,1000)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Position Over 100 Steps")
```
